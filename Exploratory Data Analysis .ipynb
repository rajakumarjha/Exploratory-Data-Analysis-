### This dataset is all about Game Of Thrones!

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

got=pd.read_csv('battles.csv')
print(got.columns)
print(got.shape)
```

```python
### this is a very clean dataset we need add any columm like title column etc.
got.head(38)
```

### Scatter Plot

```python
got.plot(kind='scatter',x='year',y='major_death')
plt.show()
```

```python
sns.set_style("whitegrid")
got.plot(kind='line',x='attacker_size',y='defender_size')
plt.show()
```

```python
#pair plot is way of reducing 3D plots to 2D
# pair plot works only when dimensionality is very less here dimentionality is 25 no way we can construct pair plot.

```

```python
# Histogram and PDF,CDF
got_Joffrey=got.loc[got["attacker_king"]=="Joffrey/Tommen Baratheon"]
got_Stark=got.loc[got["attacker_king"]=="Robb Stark"]
got_Balon=got.loc[got["attacker_king"]=="Balon/Euron Greyjoy"]
got_Baratheon=got.loc[got["attacker_king"]=="Stannis Baratheon"]
#Attaker size of each
plt.plot(got_Joffrey["attacker_size"],np.zeros_like(got_Joffrey['attacker_size']))
plt.plot(got_Stark["attacker_size"],np.zeros_like(got_Stark['attacker_size']))
plt.plot(got_Balon["attacker_size"],np.zeros_like(got_Balon['attacker_size']))
plt.plot(got_Baratheon["attacker_size"],np.zeros_like(got_Baratheon['attacker_size']))
```

```python
#defender size of each
plt.plot(got_Joffrey["defender_size"],np.zeros_like(got_Joffrey['defender_size']))
plt.plot(got_Stark["defender_size"],np.zeros_like(got_Stark['defender_size']))
plt.plot(got_Balon["defender_size"],np.zeros_like(got_Balon['defender_size']))
plt.plot(got_Baratheon["defender_size"],np.zeros_like(got_Baratheon['defender_size']))
```

```python
#univariate analysis
#defender size of each king
sns.FacetGrid(got,hue="attacker_king",size=8).map(sns.distplot,"defender_size").add_legend()
# this is a histogram + pdf of exploring defending army of each king
# histogram represent the frquency of a paticular data 
# this histogram tells us : Baratheon has highest defender army
```

```python
#attacker size of each king
sns.FacetGrid(got,hue="attacker_king",size=15).map(sns.distplot,"attacker_size").add_legend()
# this histogram shows that :  Baratheon has hightest attcking army
```

```python
# cdf is also a 1D representation of data : we can draw CDF for better accuracy
```

```python
#Mean,Varaince,std-deviation
print("1.Attackeing army of Joffrey")
print("Mean :",np.mean(got_Joffrey["attacker_size"]))
print("Max :",np.max(got_Joffrey["attacker_size"]))
print("Min :",np.min(got_Joffrey["attacker_size"]))
print("Median:",np.median(got_Joffrey["attacker_size"]))
print("std :",np.std(got_Joffrey["attacker_size"]))
print("2.Attackeing army of Stark")
print("Mean:",np.mean(got_Stark["attacker_size"]))
print("Max :",np.max(got_Stark["attacker_size"]))
print("Min :",np.min(got_Stark["attacker_size"]))
print("Median:",np.median(got_Stark["attacker_size"]))
print("std :",np.std(got_Stark["attacker_size"]))
print("3.Attackeing army of Balon")
print("Mean:",np.mean(got_Balon["attacker_size"]))
print("Max :",np.max(got_Balon["attacker_size"]))
print("Min :",np.min(got_Balon["attacker_size"]))
print("Median:",np.median(got_Balon["attacker_size"]))
print("std :",np.std(got_Balon["attacker_size"]))
print("4.Attackeing army Baratheon")
print("Mean:",np.mean(got_Baratheon["attacker_size"]))
print("Max :",np.max(got_Baratheon["attacker_size"]))
print("Min :",np.min(got_Baratheon["attacker_size"]))
print("Median:",np.median(got_Baratheon["attacker_size"]))
print("std :",np.std(got_Baratheon["attacker_size"]))
```

```python
# on an average attacking army used by Baratheon was much larger than any other king
# Balon has used much smaller army
```

```python
print("1.Defendeing army of Joffrey")
print("Mean :",np.mean(got_Joffrey["defender_size"]))
print("Max :",np.max(got_Joffrey["defender_size"]))
print("Min :",np.min(got_Joffrey["defender_size"]))
print("Median:",np.median(got_Joffrey["defender_size"]))
print("std :",np.std(got_Joffrey["defender_size"]))
print("2.Defendeing army of Stark")
print("Mean:",np.mean(got_Stark["defender_size"]))
print("Max :",np.max(got_Stark["defender_size"]))
print("Min :",np.min(got_Stark["defender_size"]))
print("Median:",np.median(got_Stark["defender_size"]))
print("std :",np.std(got_Stark["defender_size"]))
print("3.Defendeing army of Balon")
print("Mean:",np.mean(got_Balon["defender_size"]))
print("Max :",np.max(got_Balon["defender_size"]))
print("Min :",np.min(got_Balon["defender_size"]))
print("Median:",np.median(got_Balon["defender_size"]))
print("std :",np.std(got_Balon["defender_size"]))
print("4.Defendeing army Baratheon")
print("Mean:",np.mean(got_Baratheon["defender_size"]))
print("Max :",np.max(got_Baratheon["defender_size"]))
print("Min :",np.min(got_Baratheon["defender_size"]))
print("Median:",np.median(got_Baratheon["defender_size"]))
print("std :",np.std(got_Baratheon["defender_size"]))
```

```python
# On an average Stark has used more defender army than baeatheon
# On an average baeatheon has used more army than other. It seems Baratheon might have a great army hence might be a great emperor
# Stark has emphasise more on defending side than other . It seems he is clever and loves dear ones.
```

```python
# percentile is a great tool it gives us information about how much data is there that is greater than that particular selected data.
# if a data in asorted list has 90th percentile it means 10% of data are more than this value.
print("1.Defendeing army of Joffrey")
print("percentile :",np.percentile(got_Joffrey["defender_size"],np.arange(0,100,25)))
print("2.Defendeing army of Stark")
print("percentile:",np.percentile(got_Stark["defender_size"],np.arange(0,100,25)))
print("3.Defendeing army of Balon")
print("percentile:",np.percentile(got_Balon["defender_size"],np.arange(0,100,25)))
print("4.Defendeing army Baratheon")
print("percentile:",np.percentile(got_Baratheon["defender_size"],np.arange(0,100,25)))
print("1.Attackeing  army of Joffrey")
print("percentile :",np.percentile(got_Joffrey["attacker_size"],np.arange(0,100,25)))
print("2.Attackeing  army of Stark")
print("percentile:",np.percentile(got_Stark["attacker_size"],np.arange(0,100,25)))
print("3.Attackeing  army of Balon")
print("percentile:",np.percentile(got_Balon["attacker_size"],np.arange(0,100,25)))
print("4.Attackeing  army Baratheon")
print("percentile:",np.percentile(got_Baratheon["attacker_size"],np.arange(0,100,25)))
```

```python
# median Absolute Deviation is like standard deviation only calculated over median.
# MAD is not affected by outliers
from statsmodels import robust
print("1.Army of Joffrey")
print("MAD of defender :",robust.mad(got_Joffrey["defender_size"]))
print("MAD of attacker :",robust.mad(got_Joffrey["attacker_size"]))
print("2.Army of Stark")
print("MAD of defender :",robust.mad(got_Stark["defender_size"]))
print("MAD of attacker :",robust.mad(got_Stark["attacker_size"]))
print("3.Army of Balon")
print("MAD of defender :",robust.mad(got_Balon["defender_size"]))
print("MAD of attacker:",robust.mad(got_Balon["attacker_size"]))
print("4.Army Baratheon")
print("MAD of defender :",robust.mad(got_Baratheon["defender_size"]))
print("MAD of attacker :",robust.mad(got_Baratheon["attacker_size"]))
```

```python
# Box plot with whiskers : Another method of visualizing the 1-D scatter plot.
# IQR: Inter qantile Range . As name suggest it describes the same. Eg: 75th percentile-25th Percentile = c than in short range
# of C 50% of our data is residing. 
# Box plot uses same concept
print("Boxplot")

sns.boxplot(x="attacker_king",y="defender_size",data=got,width=0.8)
# line that appears is indicating 50 percentile
```

```python
sns.boxplot(x="attacker_king",y="attacker_size",data=got,width=0.7,)
```

```python
# Violin Plots: A combination of Boxplot and Whisker

sns.violinplot(x="attacker_king",y="attacker_size",data=got,width=0.4)
```

```python
sns.violinplot(x="attacker_king",y="defender_size",data=got,width=0.8)
# that white point shows 50 percentile positon, above to which is 75% and below to which is 25%
```

```python
#multivariate probability density, countour plot
sns.jointplot(x="defender_size",y="attacker_size",data=got_Baratheon,kind="kde")
# between aprxx 400 to 7000 the density is high for defender
# between aprxx 500 to 15000 density is high for attacker
```
